```python
import random

alphabet = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n","o","p","q","r","s","t","u","w","x","y","z"]
random.shuffle(alphabet)

forward = {"a": alphabet[0], "b": alphabet[1], "c": alphabet[2], "d": alphabet[3], "e": alphabet[4], "f": alphabet[5], "g": alphabet[6], "h": alphabet[7], "i": alphabet[8], "j": alphabet[9], "k": alphabet[10], "l": alphabet[11], "m": alphabet[12], "n": alphabet[13],"o": alphabet[14],"p": alphabet[15],"q": alphabet[16],"r": alphabet[17],"s": alphabet[18],"t": alphabet[19],"u": alphabet[20],"w": alphabet[21],"x": alphabet[22],"y": alphabet[23],"z": alphabet[24]}
def find_key(v): 
    for k, val in forward.items(): 
        if v == val: 
            return k 
    return "pas de clé"

def codage(string):
    newString = ""
    for letter in string:
        newString = newString + forward[letter]
    return newString

cod = codage("chien")
print(cod)

def decodage(string):
    newString = ""
    for letter in string:
        newString = newString + find_key(letter)
    return newString

decod = decodage(cod)
print(decod)
```

```python
#calcul de 2*(prod des k = 1 à inf de 4k2/ (4k2 - 1))

def prodNaif(n):
    number = 4/3
    for k in range(2, n):
        K = 4*k**2 
        number  = number * (K/(K-1))
    number = number*2
    return number

def prod(n):
    array = np.arange(1, n)
    array = 4*array**2 / (4*array**2 - 1)
    return 2*np.prod(array)
    
n=10000
print(prodNaif(n), prod(n))
%timeit prodNaif(n)
%timeit prod(n)
```

```python
def quicksort(ll):

    if len(ll) <= 1:
        return ll
    else:
        pivot = ll.pop(-1)
        less = []
        greater = []
        for x in ll:
            if x <= pivot:
                less.append(x) # append 'x' to 'less'
            else:
                greater.append(x) # append 'x' to 'greater'
        pivotList = []
        pivotList.append(pivot)
        sort = quicksort(less) + pivotList + quicksort(greater)
        return sort  # concatenate quicksort(less), pivot and quicksort(greater)
    
    
quicksort([-2, 3, 5, 1, 3])
```

```python
M = np.random.uniform(-1, 1, (5, 20))
                              
G = np.dot(np.transpose(M), M)

G_sym = (G + np.transpose(G))/2

print(np.all(np.equal(G, G_sym)))

val_propre = np.linalg.eig(G)

val_propre_reel = np.array([])
for val in val_propre:
    val_propre_reel =np.append(val_propre_reel, np.real(val))

print(np.all(np.logical_not(val_propre_reel < 0)))

print(np.linalg.matrix_rank(G))

```
