```python
import numpy as np
import pandas as pd
import matplotlib as plt
import urllib.request
import datetime as dt
import calendar as cld
```

```python
# On définit les heures d'arrivées cycliques constatées dans la base de données des classements
tuple = (8, 4, 21, 17, 14, 11)
```

```python
# On définit la fonction permettant de charger les classements

def charger_classements(annee, mois):
    cal = cld.Calendar()
    for x in cal.itermonthdays3(annee, mois):
        for r in range(6):
            try :
                url = f'https://www.vendeeglobe.org/download-race-data/vendeeglobe_{x[0]}{"%02d" % x[1]}{"%02d" % x[2]}_{"%02d" % tuple[r]}0000.xlsx'
                local_filename, headers = urllib.request.urlretrieve(url, filename=f'./Classements/vendeeglobe_{x[0]}{"%02d" % x[1]}{"%02d" % x[2]}_{"%02d" % tuple[r]}0000.xlsx', reporthook=None, data=None)
                html = open(local_filename)
                html.close()
                print(url, '            ____DOWNLOADED')

            except OSError:
                print(f'https://www.vendeeglobe.org/download-race-data/vendeeglobe_{x[0]}{"%02d" % x[1]}{"%02d" % x[2]}_{"%02d" % tuple[r]}0000.xlsx',    '            ____URL_NOT_FOUND')
```

```python
charger_classements(2020, 11)
charger_classements(2020, 12)
charger_classements(2021, 1)
charger_classements(2021, 2)
charger_classements(2021, 3)
```
